```python
labels = torch.load("all_possible_labels.pt").cuda()
TOTAL = len(labels)
ll = torch.randn(TOTAL, 512).cuda()
latents = [(ll, model.GAN.G.num_layers)]
rows = np.sqrt(TOTAL).astype(int)+1

model.set_evaluation_parameters(latents_to_evaluate=latents, labels_to_evaluate=labels, reset=True, total=TOTAL)
average_generated_images = model.evaluate(use_mapper=True, truncation_trick=1.4, only_ema=True)

# torchvision.utils.save_image(generated_iamges, 'generated_images.png', nrow=rows)
torchvision.utils.save_image(average_generated_images, 'average_generated_images.png', nrow=rows)
```

```python
"""
    0    |      1     |    2   |        3        |        4       |        5        |    6     |   7   |  8  
 --------|------------|--------|-----------------|----------------|-----------------|----------|-------|----- 
  angry  | disgust    | fear   | happy           | neutral        | sad             | surprise |       |     
  0-2    | 10-19      | 20-29  | 3-9             | 30-39          | 40-49           |  50-59   | 60-69 | 70+ 
  Female | Male       |        |                 |                |                 |          |       |     
  Black  | East Asian | Indian | Latino Hispanic | Middle Eastern | Southeast Asian | White    |       |     
  Asian  | Black      | Indian | White           |                |                 |          |       |     

""""""                 0  1  2  3  4  5  6  7  8   """
label1 = torch.Tensor([1, 0, 0, 0, 0, 0, 0,        # Expression
                       0, 0, 0, 0, 0, 0, 0, 0, 0,  # Age
                       0, 1,                       # Gender
                       0, 0, 0, 0, 0, 0, 0,        # Race7
                       0, 0, 0, 1])                # Race4
lines_to_generate = 8
TOTAL = 8
labels = label1.repeat(TOTAL, 1).cuda()
ra = {0:7, 1:10, 2:8, 3:9, 4:11, 5:12, 6:13, 7:14, 8:15}

# ll = torch.load("latents.pt")
ll = torch.randn(TOTAL, 512).cuda()
latents = [(ll, model.GAN.G.num_layers)]
```

```python
start = time.time()
test_image = None
for i in range(lines_to_generate):

    labels[:,ra[i]] = 1
    model.set_evaluation_parameters(latents_to_evaluate=latents, labels_to_evaluate=labels, reset=True, total=TOTAL)
    average_generated_images = model.evaluate(use_mapper=True, truncation_trick=1., only_ema=True)

    test_image = torch.cat((test_image, average_generated_images)) if i != 0 else average_generated_images
    labels[:,ra[i]] = 0
end = time.time()    
print("Total elapsed time: %.2f seconds" % (end - start))
print("Avg time per image: %.10f seconds" % ((end - start)/(TOTAL*lines_to_generate)))
torchvision.utils.save_image(test_image, f'test.png', nrow=TOTAL)
plt.figure(figsize=(TOTAL*2 if TOTAL < 4 else 8,lines_to_generate*2 if lines_to_generate < 4 else 8))
plt.axis("off")
plt.imshow(plt.imread("test.png"));
```

```python
os.makedirs(f'training_progress_{name}') if ~os.path.exists(f'training_progress_{name}') else None
for m in tqdm(range(len(os.listdir(os.path.join(root, name))))):
    model.load(m, root=root)
    test_image = None
    for i in range(lines_to_generate):

        labels[:,ra[i]] = 1
        model.set_evaluation_parameters(latents_to_evaluate=latents, labels_to_evaluate=labels, reset=True, total=TOTAL)
        generated_images, average_generated_images = model.evaluate(use_mapper=True, truncation_trick=1.5)

        test_image = torch.cat((test_image, average_generated_images)) if i != 0 else average_generated_images
        labels[:,ra[i]] = 0

    torchvision.utils.save_image(test_image, f'training_progress_{name}/{str(m).zfill(3)}.png', nrow=TOTAL)
```

# FID de los modelos

```python
m, step = 0, 1
fids = pd.read_csv(os.path.join(root, name, "metrics_test_all_2_100k.csv")).dropna()
fids_m = fids.sort_index()[fids.index > m][::step].reset_index()
ax = fids_m.plot.scatter("recall", "precision", color="b", figsize=(5,5))
highPareto = True

'''Pareto frontier selection process'''
sorted_list = sorted([[fids_m.recall[i], fids_m.precision[i]] for i in range(len(fids_m.recall))], reverse=True)
pareto_front = [sorted_list[0]]
for pair in sorted_list[1:]:
    if highPareto:
        if pair[1] >= pareto_front[-1][1]:
            pareto_front.append(pair)
    else:
        if pair[1] <= pareto_front[-1][1]:
            pareto_front.append(pair)

'''Plotting process'''
pf_X = [pair[0] for pair in pareto_front]
pf_Y = [pair[1] for pair in pareto_front]
plt.plot(pf_X, pf_Y, color="orange")

fids_m[fids_m.fid == fids_m.fid.min()].plot.scatter("recall", "precision", ax=ax, color="red")
fids_m[fids_m.f1 == fids_m.f1.max()].plot.scatter("recall", "precision", ax=ax, color="yellow")

mplcursors.cursor(ax).connect("add", lambda sel: sel.annotation.set_text(
    f"Modelo: {fids_m.index[sel.index]+m+1}\nFID: {fids_m.iloc[sel.index].fid}\n" + 
    f"Precision: {fids_m.iloc[sel.index].precision:.2f}\nRecall: {fids_m.iloc[sel.index].recall}\n" + 
    f"F1: {fids_m.iloc[sel.index].f1}"
) if isinstance(sel.artist, mpl.collections.PathCollection) else sel.annotation.set_text(None))
ax.set_ylim(0,1)
ax.set_xlim(0,1)
# ax.set_xlim(610,800)
```
