### Interpolation

```python
def interpolate_points(p1, p2, n_steps=10):
	# interpolate ratios between the points
	ratios = np.linspace(0, 1, num=n_steps)
	# linear interpolate vectors
	v = (1.0 - ratios[0]) * p1 + ratios[0] * p2
	vectors = torch.Tensor(v.cpu())
	for ratio in ratios[1:]:
		v = (1.0 - ratio) * p1 + ratio * p2
		vectors = torch.vstack((vectors,v.cpu()))
	return vectors.cuda()

steps = 60
label1 = torch.Tensor([0, 0, 0, 0, 0, 0, 0,        # Expression
                       0, 0, 0, 0, 1, 0, 0, 0, 0,  # Age
                       1, 0,                       # Gender
                       1, 0, 0, 0, 0, 0, 0,        # Race7
                       0, 1, 0, 0])                # Race4
labels = label1.repeat(7, 1).cuda()

for i in range(7): labels[i,i] = 1

vecs = interpolate_points(labels[0], labels[1], n_steps=steps)
for i in range(1,6):
	vecs = torch.cat((vecs, interpolate_points(labels[i], labels[i+1], n_steps=steps)), dim=0)

ll = torch.randn(1, 512).repeat(vecs.shape[0],1).cuda()
latents = [(ll, model.GAN.G.num_layers)]


model.set_evaluation_parameters(latents_to_evaluate=latents, labels_to_evaluate=vecs, reset=True, total=vecs.shape[0])
average_generated_images = model.evaluate(use_mapper=True, truncation_trick=1.4, only_ema=True)
# os.makedirs(f'training_progress_{name}') if ~os.path.exists(f'training_progress_{name}') else None
plt.figure()
plt.imshow(torchvision.utils.make_grid(average_generated_images.cpu(), nrow=19, padding=0).permute(1,2,0).numpy())
plt.show()

```

```python
clean = False
colors = ["green", "green", "lime", 
          "purple", "purple", "magenta",
          "maroon", "maroon", "red",
          "orange", "orange", "yellow",
          "royalblue", "royalblue", "cyan"]
ax = plt.figure().add_subplot(111)
for i, [root, name, label, end] in enumerate(zip(["final_models", "models", "models", "models", "models"], 
                                            ["test_all_2_100k", "test_all_4_100k", "test_all_4_100k", "test_all_5_100k", "test_all_5_100k"], 
                                            ["128_cs", "256_ms", "256_ms_wl", "128_ss", "128_ss_wl"],
                                            ["", "", "_with_labels", "", "_with_labels"])):
    m, step = 1, 1
    fids = pd.read_csv(os.path.join(root, name, f"metrics_{name}{end}.csv")).dropna()
    fids_m = fids.sort_index()[::step].reset_index()
    fids_m = fids_m[fids_m.f1 >= fids_m.f1.mean()].reset_index() if clean else fids_m

    fids_m.plot.scatter("recall", "precision", color=colors[i*3], ax=ax)
    highPareto = True

    '''Pareto frontier selection process'''
    sorted_list = sorted([[fids_m.recall[i], fids_m.precision[i]] for i in range(len(fids_m.recall))], reverse=True)
    pareto_front = [sorted_list[0]]
    for pair in sorted_list[1:]:
        if highPareto:
            if pair[1] >= pareto_front[-1][1]:
                pareto_front.append(pair)
        else:
            if pair[1] <= pareto_front[-1][1]:
                pareto_front.append(pair)

    '''Plotting process'''
    pf_X = [pair[0] for pair in pareto_front]
    pf_Y = [pair[1] for pair in pareto_front]
    plt.plot(pf_X, pf_Y, color=colors[i*3+1], path_effects=[pe.Stroke(linewidth=3, foreground='black'), pe.Normal()])

    # fids_m[fids_m.fid == fids_m.fid.min()].plot.scatter("recall", "precision", ax=ax[i], color="red")
    best_f1 = fids_m[fids_m.f1 == fids_m.f1.max()]
    best_f1.plot.scatter("recall", "precision", ax=ax, color=colors[i*3+2], 
                                                       edgecolors='black', s=80, marker="p", label=f"Mejor F1 ({name.split('_')[2]}) {label}", zorder=10)
    plt.text(best_f1.recall.values[0] + 0.0025, best_f1.precision.values[0] + 0.0025, 
             f"Modelo {best_f1.model.values[0]}", color="black", fontsize=8)                                                       
    # mplcursors.cursor(ax).connect("add", lambda sel: sel.annotation.set_text(
    #     f"Modelo: {fids_m.iloc[sel.index].model}\nFID: {fids_m.iloc[sel.index].fid}\n" + 
    #     f"Precision: {fids_m.iloc[sel.index].precision}\nRecall: {fids_m.iloc[sel.index].recall}\n" + 
    #     f"F1: {fids_m.iloc[sel.index].f1}"
    # ) if isinstance(sel.artist, mpl.collections.PathCollection) else sel.annotation.set_text(None))
    ax.legend(loc=[0.2, 0.05])
ax.set_title("Precision y recall")
# ax.set_xlim(0.0, 0.175)
# ax.set_ylim(0.175, 0.36);
```
