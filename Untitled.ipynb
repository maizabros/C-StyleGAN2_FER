```python
import os
import json
import fire
import torch
import matplotlib.pyplot as plt
from trainer import Trainer
import torchvision
import time
import cv2
import numpy as np
import glob
from tqdm import tqdm
```

```python
root = 'models'
name = "test_all_2_100k"

with open(os.path.join(root, name, 'config.json'), 'r') as file:
    config = json.load(file)
model = Trainer(**config)
model.load(-1, root=root)
```

```python
latents = [(torch.randn(TOTAL, 512).cuda(), model.GAN.G.num_layers)]
```

```python
"""
    0    |      1     |    2   |        3        |        4       |        5        |    6     |   7   |  8  
 --------|------------|--------|-----------------|----------------|-----------------|----------|-------|----- 
  angry  | disgust    | fear   | happy           | neutral        | sad             | surprise |       |     
  0-2    | 10-19      | 20-29  | 3-9             | 30-39          | 40-49           |  50-59   | 60-69 | 70+ 
  Female | Male       |        |                 |                |                 |          |       |     
  Black  | East Asian | Indian | Latino Hispanic | Middle Eastern | Southeast Asian | White    |       |     
  Asian  | Black      | Indian | White           |                |                 |          |       |     

""""""                 0  1  2  3  4  5  6  7  8   """
label1 = torch.Tensor([0, 0, 0, 1, 0, 0, 0,        # Expression
                       0, 0, 0, 0, 0, 0, 0, 0, 0,  # Age
                       0, 1,                       # Sex
                       0, 0, 0, 0, 0, 0, 1,        # Race7
                       0, 0, 0, 1])                # Race4
lines_to_generate = 8
TOTAL = 8
labels = label1.repeat(TOTAL, 1).cuda()
ra = {0:7, 1:10, 2:8, 3:9, 4:11, 5:12, 6:13, 7:14, 8:15}
```

```python
start = time.time()
test_image = None
for i in range(lines_to_generate):

    labels[:,ra[i]] = 1
    model.set_evaluation_parameters(latents_to_evaluate=latents, labels_to_evaluate=labels, reset=True, total=TOTAL)
    generated_images, average_generated_images = model.evaluate(use_mapper=True, truncation_trick=1.42)

    test_image = torch.cat((test_image, average_generated_images)) if i != 0 else average_generated_images
    labels[:,ra[i]] = 0
end = time.time()    
print("Total elapsed time: %.2f seconds" % (end - start))
print("Avg time per image: %.10f seconds" % ((end - start)/(TOTAL*lines_to_generate)))
torchvision.utils.save_image(test_image, f'test.png', nrow=TOTAL)
plt.figure(figsize=(TOTAL*2 if TOTAL < 4 else 8,lines_to_generate*2 if lines_to_generate < 4 else 8))
plt.axis("off")
plt.imshow(plt.imread("test.png"));
```

```python
for m in tqdm(range(400)):
    model.load(m, root=root)
    test_image = None
    for i in range(lines_to_generate):

        labels[:,ra[i]] = 1
        model.set_evaluation_parameters(latents_to_evaluate=latents, labels_to_evaluate=labels, reset=True, total=TOTAL)
        generated_images, average_generated_images = model.evaluate(use_mapper=True, truncation_trick=1.5)

        test_image = torch.cat((test_image, average_generated_images)) if i != 0 else average_generated_images
        labels[:,ra[i]] = 0

    torchvision.utils.save_image(test_image, f'training_progress/{str(m).zfill(3)}.png', nrow=TOTAL)
```

```python
img_array = []
for filename in glob.glob('training_progress/*.png'):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
```

```python
out = cv2.VideoWriter('training_progress/training.mp4',cv2.VideoWriter_fourcc(*'DIVX'), 30, size)
 
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
